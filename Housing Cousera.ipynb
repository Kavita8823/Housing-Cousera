# Hypothesis

1. Is there significant difference in median value of houses bounded by the Charles river or not?

H0: there is no difference in median value of houses bounded by the Charles river

H1 : there is difference in median value of houses bounded by the Charles river

alpha = 0.05 
(T-test for independent samples)

```python
import scipy.stats
```

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 1]['MEDV'],
                   boston_df[boston_df['CHAS'] == 0]['MEDV'], equal_var = True)
```

p=value < alpha ==>reject H0
==> there is difference in median value of houses bounded by the Charles river

2. Is there a difference in MEdian values of houses (MEDV) for each proportion of owner occupied units built prior to 1940?

ANOVA

𝐻0: µ1=µ2=µ3  (the three population means are equal)
𝐻1:  At least one of the means differ

```python
thrityfive_lower = boston_df[boston_df['age_group'] == '35 years and younger']['MEDV']
thrityfive_seventy = boston_df[boston_df['age_group'] == 'between 35 and 70 years']['MEDV']
seventy_older = boston_df[boston_df['age_group'] == '70 years and older']['MEDV']
```

```python
f_statistic, p_value = scipy.stats.f_oneway(thrityfive_lower, thrityfive_seventy, seventy_older)
print("F_Statistic: {0}, P-Value: {1}".format(f_statistic,p_value))
```

p=value < alpha ==>reject H0
==> we will reject the null hypothesis as there is significant evidence that at least one of the means differ.

3. Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town?

Pearson Correlation

𝐻0:  Nitric oxide concentrations score is not correlated with proportion of non-retail business acres per town
𝐻1:  Nitric oxide concentrations score is correlated with proportion of non-retail business acres per town

```python
scipy.stats.pearsonr(boston_df['INDUS'], boston_df['NOX'])
```

```python
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

p=value < alpha ==>reject H0
==> we will reject the null hypothesis as there is significant evidence that at Nitric oxide concentrations score is correlated with proportion of non-retail business acres per town

4. What is the impact of an additional weighted distance to the five Boston employment centres on the meadian value of owner occupied home?

𝐻0:β1  = 0 (additional weighted distance to the five Boston employment centres has no effect on the meadian value of owner occupied home)
𝐻1:β1  is not equal to 0 (additional weighted distance to the five Boston employment centres has effect on the meadian value of owner occupied home)

```python
import statsmodels.api as sm
```

```python
## X is the input variables (or independent variables)
X = boston_df['DIS']
## y is the target/dependent variable
y = boston_df['MEDV']
## add an intercept (beta_0) to our model
X = sm.add_constant(X) 

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

# Print out the statistics
model.summary()
```

less than the alpha (α) level = 0.05, so we reject the null hypothesis as there is evidence that there is a difference in mean evaluation scores based on gender.
